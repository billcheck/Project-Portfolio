### - <font color=blue>Target Variable vs Continuous Features</font> -

### <font color=purple>*Independent T-Tests: Continuous Feature Values and their respective Churn Categorical Values*</font>

When performing Exploratory Data Analysis on large feature sets I like to take a statistical approach in the beginning to set the direction. In the spirit of this, we will start by performing an Independent T-Test of all Continuous Features by splitting their values on our target's categorical values and testing whether or not having a specific continuous value affects the probability of user churn.

```python
# Conduct Independent T-Test.
# Does user churn affect the probability of being part of a specific category
# Ho: Observed Values are from the Population. Difference is due to Random Variation

# Create empty dataframe
ttest_results = pd.DataFrame(columns=['Feature','t-Stat','p-value'])

# Populate dataframe with T-test results
unique_cats = DRV_Jan2016["is_churn"].unique()
grouped_df = DRV_Jan2016.groupby("is_churn")
for feature in all_cont_feats:
    for i in range(0, len(unique_cats)):
        for j in range(i+1, len(unique_cats)):
            x = stats.ttest_ind(DRV_Jan2016[DRV_Jan2016["is_churn"]==unique_cats[i]][feature], DRV_Jan2016[DRV_Jan2016["is_churn"]==unique_cats[j]][feature])
            ttest_results = ttest_results.append({'Feature':f'{feature}','t-Stat':f'{x.statistic}','p-value':f'{x.pvalue:1.4f}'}, ignore_index=True)
```

```python
# Print T-Test Results
ttest_results.sort_values('p-value',ascending=False).head(20)
```

Out of roughly 230 features created we have 9 features that were deemed insignificant (Level of Significance = .05). Interestingly enough, the majority of these features are trend features that capture older comparative data. This lead's me to assume that activity closer to the member's expiration date has more influence on churn. Let's look are both significant and insignificant distributions through box plots.

```python
# Make List on non-significant features
cont_feats_nonsig = ttest_results.sort_values('p-value',ascending=False).head(9)['Feature']
cont_feats_nonsig
```

### <font color=purple>*Chi_Square Test of Independence*</font>

When analyzing Categorical Features, we can dig into their respective contingency table values for Churn @ 1 and 0 and test if they are of the same distribution. Here I am looking to understand  if being part of a specific category affects the probability of user churn.

```python
# Conduct Chi-square Test of Independence.
# Does user churn affect the probability of being part of a specific category
# Ho: Categories are not Independent. Difference is due to Random Variation
from scipy.stats import chi2_contingency
import numpy as np

# Create empty dataframe
chi2_result = pd.DataFrame(columns=['Feature','Chi2-Stat','P-Value'])

# Create Contingency Table and Run test
for feature in all_cat_feats:
    contingency_table=pd.crosstab(DRV_Jan2016["is_churn"],DRV_Jan2016[feature])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_result = chi2_result.append({'Feature':f'{feature}','Chi2-Stat':f'{chi2}','P-Value': f'{p:1.4f}'}, ignore_index=True)

print(f'''
{chi2_result.sort_values('P-Value', axis=0, ascending=False)}\n
''')
```

We do not have many Categorical Features, but with such low P-Values we can see that all these features are statistically significant. Let's take a deeper look into a couple of them.

### - <font color=blue>Create Balanced Dataset</font> -

To avoid any class imbalance issues we will create a balanced set for model training purposes.

```python
# Query to create balanced dataset on BigQuery
for x in [1,3,5,7,9,11,13]:
    QUERY = f"""
                CREATE TABLE `spark-kkbox.KKbox_User_Data.DRV_Jan2016_{x}to1` AS
                 (SELECT 
                    *
                  FROM 
                    `spark-kkbox.KKbox_User_Data.DRV_Jan2016`
                 
                 WHERE 
                    is_churn = 0
                  ORDER BY 
                    RAND()
                  LIMIT 
                    {18865*x})

                UNION ALL

                  SELECT 
                    *
                  FROM 
                    `spark-kkbox.KKbox_User_Data.DRV_Jan2016`
                  WHERE 
                    is_churn = 1 
             """
    # Call .query() followed by .result() to trigger the 'lazy function'
    bigquery_client.query(QUERY).result()
```
