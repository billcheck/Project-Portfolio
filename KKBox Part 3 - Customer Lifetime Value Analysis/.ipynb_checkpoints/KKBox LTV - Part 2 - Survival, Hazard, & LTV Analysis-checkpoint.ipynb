In the previous projects we discussed how beyond prediction, customer retention inevitably comes down to understanding Customer Value and the price we are willing to pay to keep this value. 1) Which Customers Have the Highest Probability of Churning Soon?

In addressing question #1, we developed a churn prediction model which provided several strategies that we could leverage in identifying high risk customers/members. **Lifetime Value (LTV) and Why is it important?**

LTV, aka Value of Acquisition and Retention, is defined as the present value of the future (net) cash flows. This is a forward-looking concept and is not a measure of past profitability. Additionally, it also gives us the invaluable ability to identify which channels are producing our most valuable customers (via attribution). In this project we will dive into its calculation and discuss its implications for our business. Let's get started.

## 1) Survival Analysis

Addressing questions such as ‘how long would it be, before a particular event occurs’, Survival Analysis is a branch of statistics that was originally used in medical and biological disciplines to study the survival times of patients suffering from major diseases. We refer to this a ‘time to event’ analysis. This analysis can be further applied to not just traditional death events, but to many different types of events of interest in different business domains. With the help of Survival Analysis we can focus on churn prevention efforts of high-value customers with low survival time.

Keep in mind however that this is more of an EDA-technique to evaluate the expected lifetime of different cohort groups. Similarly to how we would conduct a bi-variate EDA to understand the behavior and effect of individual features before creating a regression model, we do the same thing here by conducting a Survival Analysis before creating Cox Proportional Hazards (CPH) Regression. Let’s now see what information we can gather using a specific type of survival curves called Kaplan-Meier Curve.


### <font color=purple>Overall Population</font>

First we will take an overall observation of all users to get the big picture of what is the underlying behavior. 

```python
## Time to event data of censored and event data
T = CLV_Feb2016['tenure']

## It has the churned (1) and censored is (0)
E = CLV_Feb2016['is_churn'].apply(lambda x: True if x == 1 else False)

# Instantiate Model
kmf = KaplanMeierFitter()

## Fit the data into the model
kmf.fit(T, E, label='Kaplan Meier Estimate')

## Create an estimate
plt.figure(figsize=(10,5))
plt.title('Overall Population', fontsize='xx-large')
kmf.plot()
```

Overall retention seems pretty strong with ***> 95%*** probability of retaining users past 400 days. **We do observe a drop off but I don't know why that is present.** Overall it looks like KKBox is quite a healthy business as the odds of a customer surviving are high. Let's break up our data into various cohorts to develop a solid foundation before determining Lifetime Value for our customers. It is important note that the variables which we are creating our cohorts off of do not and cannot change over time.

## 2) Cox Proportional Hazard Model (Survival Regression)

As previously discussed, our Kaplan Meier curves work with only one categorical predictor ignoring the effects of all other predictors. Hence arises a need for a better model that can simultaneously assess the impact of all the numeric and multiple predictors on survival times. In survival analysis, predictors are often referred to as covariates. To investigate the relationship between the survival time and our variables we will build a Cox Proportional Hazards (CPH) Regression.

Cox Proportional Hazards (CPH) model is a commonly used semi-parametric model used for investigating the relationship between the survival time and one or more variables (includes categorical and quantitative predictors). It returns the Expected lifetime for each user by cumulatively calculating the an individuals risk of "death" over time. ‘Hazards’ can be thought of something that would increase/decrease chances of survival. In our business problem, for example, a hazard may be the type of contract a customer has or the way they registered for the app.

One restriction is the model assumes a constant ratio of hazards over time across groups. Lifeline offers a built in check_assumptions method for the CoxPHFitter object.

### - <font color=blue>Data Pre-Processing</font> -

```python
# Specify the Continuous and Categorical features we will include in our regression model
cont_feat = ['tenure','unique_payment_plans']
cat_feat = ['registered_via','Cluster','city_agg']

# Create df for regression model
CLV_Feb2016 = CLV_Feb2016[['msno','avg_paid_per_day','is_churn'] + cont_feat + cat_feat]

# Encode Categoricals
CLV_Feb2016 = pd.get_dummies(data=CLV_Feb2016, columns=cat_feat, prefix_sep='_',drop_first=True)

# Use PCA to make 1 component for 'registered_via' to prevent multicolinearity errors
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
CLV_Feb2016['registered_via_PCA'] = pca.fit_transform(CLV_Feb2016[['registered_via_4','registered_via_7','registered_via_9','registered_via_10']]).round(decimals=3)

CLV_Feb2016 = CLV_Feb2016.drop(['registered_via_4','registered_via_7','registered_via_9','registered_via_10'], axis=1)
```

### - <font color=blue>Build CPH Regression Model via Lifelines</font> -

```python
# Instantiate a cph object
cph_lifelines = CoxPHFitter()   

# Fit the data to train the model
model_lifelines = cph_lifelines.fit(CLV_Feb2016[CLV_Feb2016.columns[2:]], duration_col='tenure', event_col='is_churn')
```

```python
model_lifelines.summary
```

In the above regression, the key output is exp(coef). This is interpreted as the scaling of hazard risk for each additional unit of the variable:
- **1        --> No Effect**
- **Below 1  --> Decrease in Hazard**
- **Above 1  --> Increase in Hazard**

Interpreting the summary

- The hazard ratio (HR) given under exp(coef) against each predictor shows that city_agg_4 and city_agg_3 feature with exp(coef) = ~.9 have a minimal effect on hazard.

- Cluster_2 and Cluster_3 features with exp(coef) = .77 and .71 respectively (i.e., less than 1) which means a decrease in hazard and hence better survival. So, a member being classified as either one of the clusters reduces the hazard by a factor of .77 and .71 respectively and thus a good prognostic for better survival.

- city_agg_1, Cluster_1, registered_via_PCA have hazard ratios = ~2 and above! These are significant "killers".

### - <font color=blue>Feature Engineering: Cumulative Hazard Prob, Lifetime, Lifetime Value</font> -

#### <font color=purple>*Current Cummulative Harzard Probability For Each User*</font>

```python
# Calculate each member's Cummulative Hazard Probability with respect to their current tenure
def cumul_hazard_prob(row):
    if row['tenure'] > 4306:
        cumul_hazard_prob = model_lifelines.predict_cumulative_hazard(row).iloc[row['tenure']-(row['tenure']-4306)]
        return list(cumul_hazard_prob)[0]
    else:
        cumul_hazard_prob = model_lifelines.predict_cumulative_hazard(row).iloc[row['tenure']]
        return list(cumul_hazard_prob)[0]
```

```python
# Calculate each member's Cummulative Hazard Probability with respect to their current tenure
start = time.time()
CLV_Feb2016['Current Hazard Prob'] = CLV_Feb2016.apply(lambda x: cumul_hazard_prob(x), axis=1)
end = time.time()
print('Time spent for training: {}'.format(round(end-start)))
```

#### <font color=purple>*Total Expected Lifetime and Lifetime Value*</font>

```python
# Calculate Total Expected Lifetime
start = time.time()
lt1 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[0:100000])
lt2 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[100000:200000])
lt3 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[200000:300000])
lt4 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[300000:400000])
lt5 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[400000:500000])
lt6 = model_lifelines.predict_expectation(CLV_Feb2016.iloc[500000:])
CLV_Feb2016['LT'] = pd.concat([lt1,lt2,lt3,lt4,lt5,lt6])
end = time.time()
print('Time spent for training: {}'.format(round(end-start)))
```

```python
# Calculate Lifetime Value
CLV_Feb2016['LTV'] = CLV_Feb2016['LT'] * CLV_Feb2016['avg_paid_per_day'] 
```

This Calculation is overly optimistic but is the effectively the Ceiling that we could theoretically achieve.

#### <font color=purple>*Remaining Lifetime*</font>

```python
# Calculate remaining LT given each member's current tenure
CLV_Feb2016['LT_remaining'] = CLV_Feb2016['LT'] - CLV_Feb2016['tenure']
```

#### <font color=purple>*Remaining Lifetime Value*</font>

Calculate NPV for this!!

```python
CLV_Feb2016['LTV_remaining'] = CLV_Feb2016['LT_remaining'] * CLV_Feb2016['avg_paid_per_day']
```

### <font color=blue>Conclusion: How can we use all of this information to improve our overall acquisition efforts?</font>

There are two main aspects of our business which we can significantly improve on given this new information: Optimizing Customer Acquisition and Activation, and also Cost of Acquisition-Retention vs Value of Acquisition-Retention.

**Optimizing Customer Acquisition and Activation**

One of the luxuries of conducting a survival analysis is that we can clearly see what are the exact factors that are contributing to both strong and weak members. In our analysis we saw how the way in which a user registered for the app was critical. Seeing this we can easily make a strong case for redistributing our resources to focus our Acquistion efforts more on those channels which are creating strong memebers. In addition to this it also showed us that we really need to evaluate why those channels are significantly under-performing to identify the root cause.

In terms of Activation, our analysis showed us that we should be pushing our users into taking the monthly plan over all other plans when they sign up. All though it is a common thought that long term plans are best for business, this analysis just made a strong case against that as they all performed significantly worse compared to their monthly counterpart. Its insights like these which make survival analyses an invaluable for all businesses. Given such a limited, and incomplete, dataset I am very impressed with the inference we have been able to determine.

**Cost of Acquisition vs Value of Acquisition**

Often when we think of customer acquisition and retention we immediately get into a “we need to minimize costs” (the floor) state of mind. Yet how is it that when we hire talent we want the best, and we are willing to pay (the ceiling) for the best? This is the exact logic that we should be applying to our Customer Acquisition and Retention efforts. CLV is 'The Ceiling' we can pay to acquire the Best Members. We shouldn’t be obsessing over pennies if there are dollars to be gained!
